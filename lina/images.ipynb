```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re
import pymongo
```

```python
list_image = []
```

```python
googlesearch_url = "https://www.google.com/search?tbm=isch&q=squash"
```

```python
response = requests.get(googlesearch_url).text
```

```python
soup = bs(response, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
result = soup.findAll('table', {"class":"images_table"})

for item in result:
    for td in item.findAll('td'):
        for br in item.findAll('cite'):
            br.decompose()
        text = td.get_text()
        text_split = text.split("-")
       # print(text_split)
        get_text = text_split[0]
        
        
        for img in td.findAll('img'):
            image = img['src']
        for link in td.findAll('a'):
            links = link['href']
            split_link = links.split("/")
            get_link = split_link[3]
            #print(get_link)
            
        
                
                
            #print('-------')
            #print(get_text)
            #print(image)
            #print(links)
            post = {
                
                "title": get_text,
                "url": image,
                "website": get_link
            }
           # print(post)
            list_image.append(post)
 
        
        
```
