```python
#Importing Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading CSV File
nutrition_df = pd.read_csv("C:\\Users\\MainUser1\\Desktop\\etl\\Nutritional_values.csv")
nutrition_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(nutrition_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,8))
plt.bar(x_axis, nutrition_df["protein"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, nutrition_df["name"], rotation="vertical")
plt.title("Protein (in mg)")
plt.xlabel("Vegetables")
plt.ylabel("Amount per 100gm")
```
