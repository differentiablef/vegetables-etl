```python
#Importing Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading CSV File
minerals_df = pd.read_csv("C:\\Users\\MainUser1\\Desktop\\etl\\minerals.csv")
minerals_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(minerals_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,8))
plt.bar(x_axis, minerals_df["calcium"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, minerals_df["name"], rotation="vertical")
plt.title("CALCIUM RICH (in mg)")
plt.xlabel("Vegetables")
plt.ylabel("Amount per 100gm")
```
